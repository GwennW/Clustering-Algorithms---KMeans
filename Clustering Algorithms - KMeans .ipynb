```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
df=make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.6,random_state=50)
```

```python
points=df[0]
```

```python
plt.scatter(df[0][:,0],df[0][:,1],c=df[1],cmap='coolwarm')
```

```python
from sklearn.cluster import KMeans
```

```python
kmean=KMeans(n_clusters=4)
```

```python
kmean.fit(points)
```

```python
pred=kmean.fit_predict(points)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(points[pred==0,0],points[pred==0,1],s=100)
plt.scatter(points[pred==1,0],points[pred==1,1],s=100)
plt.scatter(points[pred==2,0],points[pred==2,1],s=100)
plt.scatter(points[pred==3,0],points[pred==3,1],s=100)
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
dendrogram=sch.dendrogram(sch.linkage(points,method='ward'))
```

```python
hc=AgglomerativeClustering(n_clusters=4, affinity='euclidean',linkage='ward')
```

```python
pred=hc.fit_predict(points)
```

```python
plt.scatter(points[pred==0,0],points[pred==0,1],s=100)
plt.scatter(points[pred==1,0],points[pred==1,1],s=100)
plt.scatter(points[pred==2,0],points[pred==2,1],s=100)
plt.scatter(points[pred==3,0],points[pred==3,1],s=100)
```
